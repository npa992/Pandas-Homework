```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
school_num=school_data_complete['School ID'].nunique()
student_num=school_data_complete['Student ID'].nunique()

schools=school_data_complete[['School ID','budget']].drop_duplicates('School ID')
total_budget=schools['budget'].sum()

average_math_score=round(school_data_complete['math_score'].mean(),2)
average_reading_score=round(school_data_complete['reading_score'].mean(),2)
average_score=(average_math_score+average_reading_score)/2

passing_math=school_data_complete['Student ID'][(school_data_complete['math_score']>=70)].count()
passing_reading=school_data_complete[['Student ID']][(school_data_complete['reading_score']>=70)].count()
passing_overall=school_data_complete[['Student ID']][(school_data_complete['math_score']>=70)&
                                                     (school_data_complete['reading_score']>=70)].count()

district_summary=pd.DataFrame({'Number of Schools':school_num,'Number of Students':student_num,
                              'Total Budget':total_budget,'Average Math Score':average_math_score,
                              'Average Reading Score':average_reading_score,'Average Score':average_score,
                              'Passing Math':passing_math,'Passing Reading':passing_reading,
                              'Passing Overall':passing_overall})
district_summary
```

```python
school_id=school_data_complete.groupby(by='School ID')

passing_math2=school_data_complete[['School ID','math_score']][(school_data_complete['math_score']>=70)].groupby('School ID').count()


passing_reading2=school_data_complete[['School ID','math_score']][(school_data_complete['reading_score']>=70)].groupby('School ID').count()


passing_overall2=school_data_complete[(school_data_complete['math_score']>=70)&
                                      (school_data_complete['reading_score']>=70)].groupby('School ID').count()




school_name=school_data_complete[['school_name','budget']].drop_duplicates('school_name').reset_index()
school_df=pd.DataFrame({'Total Students':school_id['Student ID'].count(),
                       'School Name':school_name['school_name'],
                       'Budget':school_name['budget'],
                       'Budget per Student':school_name['budget']/school_id['Student ID'].count(),
                       'Average Math Score':school_id['math_score'].mean(),
                       'Average Reading Score':school_id['reading_score'].mean(),
                       'Percent Passing Math':passing_math2['math_score'],
                       'Percent Passing Reading':passing_reading2['math_score'],
                       'Percent Passing Overall':passing_overall2['math_score'],
                       'Pass Rate':passing_overall2['math_score']/school_id['Student ID'].count()*100})
school_df
```
